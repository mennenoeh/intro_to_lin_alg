# Totally realistic practical example: Zombie apocalypse in your analytics department

## Setting

Let's assume your analytics department consists of 100 people. On the first day, one person is a zombie and 99 people are probably up (healthy).

````python
z = 1
g = 99
````


Let's create a vector from this, call it `at_first_day` and output it. We use `ndarray` as the data type here. There are also other data types that we could use here - e.g. `matrix` or `vector`. But as just said, the data type `ndarray` is the most versatile for now. 

````python
at_first_day = np.array([z, g])
print(at_first_day)
````



    [ 1 99]
    

Zombies are of course a nasty thing. Let's assume that a healthy person has an 80% probability of being bitten by a zombie in one day and then has a 90% probability of remaining a zombie for one day. This results in the following matrix.

| | today zombie|today healthy|
|----|----|----|
|Zombie the next day|0.9|0.2|
|healthy|the next day|0.1|0.8|

Let's convert the whole thing into Python:




````python
# Probabilities per day

zz = 0.9 # remain zombie
zg = 0.1 # Zombie recovers and becomes healthy

gg = 0.8 # Healthy person remains healthy
gz = 0.2 # Healthy person is bitten by zombie

# zombie healthy today
# vv vv
zombification_matrix = np.array([[zz, gz], # &lt;- next day zombie
                                  [zg, gg]]) # &lt;- healthy the next day
print(zombification_matrix)
````

We have now transferred the probabilities to a matrix and called it a 'zombification matrix'. You may have noticed that the sum of the probabilities in each column is 1. This is an important property of this matrix, also known as the [transition matrix](https://de.wikipedia.org/wiki/Ãœbergangsmatrix) or Markov matrix. This means that the space of possibilities is complete: either you are a zombie or you are not. There is no other possibility here.

## The walking way
Let's do the math on foot! We take the variables above and multiply and add up the numbers for the next day.

````python
z_second_day = z * zz + g * gz # The zombies (z) that stay (zz) plus the healthy ones (g) that become zombies (gz).
g_second_day = g * gg + z * zg # The healthy people (g) who remain (gg) plus the zombies (z) who become healthy (zg).
print(f"On day 2 there are {z_second_day:2.4} zombies and {g_second_day:2.4} healthy ones. There are {(z_second_day/g_second_day):2.4} times more zombies than healthy people.")
````



    There are 20.7 zombies and 79.3 healthy people on day 2. There are 0.261 times more zombies than healthy people.
    

Okay, that was okay! Now we have two more problems to tackle:

1. what if we have 500 items and not 2 groups? That won't work! Meaning: The procedure **does not scale with the vector size**.
2. what does it look like after 20 days? Means: The procedure **does not scale over time**.



## The Linear Algebra Way
If you look at the calculation for the zombies the next day above, you may notice that you have come across the formula before. It is the dot product between two vectors: Namely, the vector `at_first_day` with the first row of the matrix `zombification_matrix[0,:] = [0.9 0.2]` . (In Excel, by the way, the formula is called `SUMPRODUCT()`). Let's try this out! The dot product is expressed with the data type `ndarray` with the character `@`.


````python
zombie_quotas = zombification_matrix[0,:]
print("zombie_quotas: " + str(zombie_quotas))
healthy_quotas = zombification_matrix[1,:]
print("Health quotas: " + str(healthy_quotas))
````



    Zombie odds: [0.9 0.2]
    Healthy odds: [0.1 0.8]
    




````python
z_second_day = at_first_day @ zombie_quotas
g_second_day = at_first_day @ healthy_quotas
print(f"There are on day 2 {z_second_day:2.4} zombies and {g_second_day:2.4} healthy ones. There are {(z_second_day/g_second_day):2.4} times more zombies than healthy people.")
````



    There are 20.7 zombies and 79.3 healthy people on day 2. There are 0.261 times more zombies than healthy people.
    

Hmmmmmm... We haven't really won anything yet, have we? We now have slightly more elegant code and have saved ourselves having to go through the columns of the matrix. But problem 1 and 2 are still not much help. Let's think about problem 1 first: Wouldn't it be cool if we didn't have to do this for each component of the vector individually? This is also possible with the character `@`. 

### The matrix multiplication

The `@` means **matrix multiplication**. The dot product is only a special case if both matrices have only one column or one row. We can now simply multiply the vector with the matrix and *zack* - the result is there! 

````python
at_second_day = zombification_matrix @ at_first_day
print(f"There are {at_second_day[0]:2.4} on day 2. zombies and {at_second_day[1]:2.4} healthy ones. There are {(at_second_day[0]/at_second_day[1]):2.4} times more zombies than healthy people.")
````

By the way, in Excel the formula is called `MMULT()`.

Now for the **informatician**: Pressing Enter 20 times can be automated, can't it? Let's go! The second best-known loop in the world helps us with this: the `for` loop. And because we are "real" computer scientists, we make a function out of it:


````python

def zombification(start_values, z_matrix, days: int, verbose = False):
    at_today = start_values
    result_history = []

    for i in range(days) :
        at_tomorrow = z_matrix @ at_today # Here we calculate tomorrow's zombie count.
        if verbose: # This is just so that we can display the intermediate results if required.
            print(f"There is on {i+1}. Day {at_tomorrow[0]:2.4} zombies and {at_tomorrow[1]:2.4} Healthy ones. There are {(at_tomorrow[0]/at_tomorrow[1]):2.4} times more zombies than healthy people.") # We output the count
        result_history.append(at_tomorrow) # Write away history
        at_today = at_tomorrow # We go one day into the future by setting today = tomorrow. #TomorrowIsTheNewToday 

    return np.array(result_history)

````

### Eigenstuff
#### Eigenvectors

Yup, as suspected: We don't see any change because we have directly tipped the **eigenvector** into the matrix by multiplication. So nothing happens as a result of the multiplication. This is exactly how an eigenvector is defined-ish. How would an **engineer** go about it? The cool thing is that you don't have to code all this stuff yourself. There are quite efficient methods for this. Here, for example, we use the `linalg` package from `numpy`.




````python

(eigvals, eigvectors) = LA.eig(zombification_matrix)
print(eigvectors)
````

#### Eigenvalues
You have probably already seen above that not only eigenvectors are output, but also eigenvalues. You have probably heard this word before in a PCA or factor analysis. Let's take another look at exactly what it says:


    

Okay, that's "1". What might that mean? Well, it can't be due to the starting conditions, because the result of the 'Eigen' method can only have something to do with the matrix. Let's try an experiment: What happens if we change the matrix? Let's assume that the health rate and the zombie infection each balance each other out at 20%:



````python

zombification_matrix_symm = np.array([[0.8, 0.2], # &lt;- next day zombie
                                       [0.2, 0.8]]) # &lt;- healthy the next day
print(zombification_matrix_symm)

(eigvals_symm, eigvectors_symm) = LA.eig(zombification_matrix_symm)
evec1_symm = eigvectors_symm[:, 0]
eval1_symm = eigvals_symm[0]
print(f"The eigenvector is: {evec1_symm}" )
print(f"The length of the normalized eigenvector is: {LA.norm(evec1_symm, 2)}")
print(f"The length of the eigenvector is: {LA.norm(evec1_symm*eval1_symm, 2)}")
print(f"The eigenvalue is: {eval1_symm:2.4}")
print(f"There are {evec1_symm[0]/evec1_symm[1]:2.6} times more zombies than healthy people in the long term")
````


    

Not illogical: If the zombie population is balanced, the actual zombies and healthy people will also be balanced in the long term. However, the intrinsic value remains "1". Why doesn't it change? Something must have remained the same in the matrix. Let's think about what that could be. Well, each column still results in "1", as this is a [transition matrix](https://de.wikipedia.org/wiki/Ãœbergangsmatrix). The matrix therefore keeps the division size constant with every multiplication: it does not become larger or smaller.

Let's now assume that 10% of the zombies are completely killed. So 80% of zombies remain zombies, 10% become healthy and 10% disintegrate into dust.
````python
# today zombie today healthy
# vv vv
zombification_matrix_dust = np.array([[0.8, 0.2], # &lt;- next day zombie
                                        [0.1, 0.8]]) # &lt;- healthy the next day
print(zombification_matrix_dust)
(eigvals_dust, eigvectors_dust) = LA.eig(zombification_matrix_dust)
evec1_dust = eigvectors_dust[:, 0]
eval1_dust = eigvals_dust[0]
evec1_dust_relation = evec1_dust[0]/evec1_dust[1]
print(f"The eigenvector is: {evec1_dust}" )
print(f"The length of the normalized eigenvector is: {LA.norm(evec1_dust, 2)}")
print(f"The length of the eigenvector is: {LA.norm(evec1_dust*eval1_dust, 2)}")
print(f"The eigenvalue is: {eval1_dust:2.4}")
print(f"There are {evec1_dust_relation:2.6} times more zombies than healthy people in the long term")
````

#### Excursus: Interpreter vs. compiler
This is particularly important with Python and R, because __Python and R are incredibly slow__ (video on this [here](https://www.youtube.com/watch?v=D3h62rgewZM)). The low speed is the price we have to pay so that we can execute each line (or cell) individually in Python. We need this so that we as data scientists can work iteratively on problems - e.g. in Jupyter notebooks like this one. Python is also said to be translated line by line into machine language by an __interpreter__. On the other hand, there are __compiler__ languages such as C or C++. Here, the script is translated directly and completely into machine language, which is much more efficient.

There is a super-cool (and short) video from 1983 [here](https://www.youtube.com/watch?v=_C5AHaS1mOA)
. Aged very well.

#### So how do we solve the performance problem?
JIT-compiled languages such as C# or Java are a middle ground between interpreter and compiler languages. (JIT = Just in Time). Although this can improve performance, the much-loved ability to execute lines and cells individually is lost.

It would be better if we could encapsulate individual fixed components of our calculations super-efficiently in C. And this is exactly what happens when we use ready-made packages such as Numpy or ScikitLearn! Tadaaa!

So: If there are already efficient implementations - __don't do it yourself__! But how am I supposed to know if something already exists? (unkown unknowns.) Unfortunately, the only way to find out is to go back to the mathematical basics. These are generic enough to work everywhere, well named so that you can find them and incredibly well implemented. You can also see this in the structure of this Jupyter notebook: there was already a function in numpy for almost everything that we implemented ourselves for didactic reasons.

For completeness, there are various __performance hacks__ that you can always use. A YouTube search can help here.
- Threading: Why, for example, do files not download simultaneously and not one after the other?
- Multiprocessor: Why not calculate things on several CPU cores at the same time and not one after the other?
- Caching: If a function is often called with the same parameters, the result of the function can be saved so that it does not always have to be recalculated.
- Translate entire functions into machine language (JIT compilation): This can be done with the "numba" package, for example. The first execution then takes quite a long time, but the subsequent executions are super-fast.
- Other: List comprehensions, for example, are faster than for-loops etc.

### When tables, when matrices?
Basically, tables are certainly the best way to store our data. They are the most flexible and can usually hold all the information we want to store. In principle, tables are also memory-efficient, as they do not have to store empty cells in the first place. With matrices, you have to use a different data type, the so-called sparse matrix. This is particularly useful when calculating with large matrices that are largely empty. Think of a matrix that could, for example, show which customer (rows) has watched which movie (columns) at Netflix. It will be largely empty because not every movie has been seen by every person (column view) or not every person has seen every movie (row view).

In my opinion, linear algebra is always worthwhile when "everything is somehow connected" and we are looking for the determining structures in the data. An overview of the criteria identified so far can be found in the following table.

|Property|Tables|Matrices|
|----|----|----|
|Data types|several possible - one per column|only one data type possible|
|Operations|Limited to SQL - useful for simple arithmetic operations|arithmetic operations that encapsulate complex operations, such as matrix multiplication or eigenvector calculation|
|speed|rather slow|fast|
|Flexibility|high|low|
|Storage efficiency|good for storing data, as empty cells are not stored|low if SparseMatrix is not used|
|Complexity|Sums/average values in many dimensions|Correlations between many dimensions|
